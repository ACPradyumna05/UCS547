```python
import numpy as np
```

```python
arr=np.array([1,2,3,6,5,4])
```

```python
rarr = arr[::-1]
print(rarr)
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[1,2,3],[4,5,6]])
print(arr1==arr2)
```

```python
x=np.array([1,2,3,4,5,1,2,1,1,1])
y=np.array([1,1,1,2,3,4,2,4,3,3])

mfreq_x=np.argmax(np.bincount(x))
mfreq_y=np.argmax(np.bincount(y))

print(mfreq_x, mfreq_y)
```
